```python
# setup_yolo : sets up the model to use for detecting objects in an image
def setup_yolo():
    model = YOLO("yolov8x.pt")
    return model

# detect_yolo : given an image and a model outputs the list of labels of detected ojects
def detect_yolo(image, model):
    detected = []
    results = model(image, verbose=False)
    for result in results[0].boxes:
        detected.append(int(result.cls[0]))
    return detected
```

```python
# is_url : a function to determine whether a string is a URL or not
def is_url(string):
    try:
        result = parse.urlparse(string)
        return all([result.scheme, result.netloc, result.path])
    except:
        return False

# load_image : a function to load an image
def load_image(image_path):
    if is_url(image_path):
        #return Image.open(requests.get(image_path, stream=True).raw)
        try:
          return Image.open(requests.get(image_path, stream=True).raw)
        except:
          return False
    elif os.path.exists(image_path):
        return Image.open(image_path)
    return False

# show_image : display image
def show_image(image):
  display(Img(image))
  return True
```

```python
# a function to perform inference
def get_caption(model, image_processor, tokenizer, image):
    # preprocess the image
    img = image_processor(image, return_tensors="pt").to(device)
    # generate the caption (using greedy decoding by default)
    output = model.generate(**img)
    # decode the output
    caption = tokenizer.batch_decode(output, skip_special_tokens=True)[0]
    return caption
```

```python
# classify_input: understands if input is a confirmation and return TRUE or negation return FALSE
def classify_input(input_str, start):
    inp = input_str
    # List of confirmation and negation keywords
    confirmations = {"yes", "yeah", "yep", "sure", "okay", "ok", "affirmative", "certainly", "definitely", "absolutely", "indeed", "of course", "right", "correct", "aye", "yup"}
    negations = {"no", "nah", "nope", "never", "not", "negative", "absolutely not", "by no means"}

    while True:
      if inp in confirmations:
          return True
      elif inp in negations:
          if start:
              print("[Stop chatbot: 'quit'] ChatBot: Which image would you like to talk about?")
          return False
      elif inp == "quit":
          return True

      else:
        if start:
          print("[Stop chatbot: 'quit'] ChatBot: Sorry, I couldn't understand. Is this the image you would like to talk about?")
          inp = input("You: ").lower()
        else:
          print("[Stop chatbot: 'quit'] ChatBot: Sorry, I couldn't understand. Would you like to analyze another image?")
          inp = input("You: ").lower()
```

```python
# extract_number_from_text : given a sentence extract number of objects that is been talked about
def extract_number_from_text(text):
    # Define a mapping of common phrases to numbers
    number_mapping = {
        "one": 1, "two": 2, "three": 3, "four": 4, "five": 5,
        "six": 6, "seven": 7, "eight": 8, "nine": 9, " ten": 10,
        "couple": 2, "few ": 3, "several": 3, "many": 5, "dozen": 12,
        "a": 1, "an": 1
    }

    # Check if the user input directly contains a number
    number_in_text = re.search(r'\d+', text)
    if number_in_text:
        return int(number_in_text.group())

    text = text.translate(str.maketrans('', '', string.punctuation))
    text = text.split()

    # Check for phrases in the user's text and map to numbers
    for phrase, number in number_mapping.items():
        if phrase in text:
            return number

    return None  # Return None if no number or phrase is found
```

```python
# get_synonyms : get synonyms based on the word and part of speech
def get_synonyms(word, pos):
  for synset in wn.synsets(word, pos=pos_to_wordnet_pos(pos)):
    for lemma in synset.lemmas():
        yield lemma.name()

# pos_to_wordnet_pos : return wordnet part of speech from ntlk pos
def pos_to_wordnet_pos(penntag, returnNone=False):
  morphy_tag = {'NN':wn.NOUN, 'JJ':wn.ADJ,
                'VB':wn.VERB, 'RB':wn.ADV}
  try:
      return morphy_tag[penntag[:2]]
  except:
      return None if returnNone else ''
```

### Question Generation and Evaluation

```python
# generate_questions : given a list of detected objects generate 3 questions
#                      2 question about objects not present in the image
#                      1 question about an object present in the image
def generate_questions(detected_list):
    # Find out what objects are present in the image
    present_objects = [(object_labels[obj_id], obj_id) for obj_id in detected_list if obj_id in object_labels]
    # Select one object that is present for the first question
    if present_objects:
        real_question_object = random.choice(present_objects)
    else:
        real_question_object = None

    absent_id = [ key for key in object_labels.keys() if key not in detected_list]

    # Select two objects that are not present for the other questions
    all_possible_objects = list(object_labels.values())

    if len(absent_id) >= 2:
        fake_question_objects = random.sample(absent_id, 2)
    else:
        fake_question_objects = absent_id  # fallback in case we don't have enough absent objects

    questions_list = []

    # First question: about something present
    if real_question_object:
        questions_list.append((
            f"Are there any {object_labels_plural[real_question_object[1]]} in the image? If so, how many {object_labels_plural[real_question_object[1]]} are in the image?",
            True,
            real_question_object
        ))

    # Next two questions: about things not present
    for obj in fake_question_objects:
        questions_list.append((
            f"Are there any {object_labels_plural[obj]} in the image? If so, how many {object_labels_plural[obj]} are in the image?",
            False,
            (object_labels[obj], obj)
        ))

    # Shuffle questions to randomize order
    random.shuffle(questions_list)

    return questions_list
```

```python
# check_answer : given questions, answers and detected objects
#                that checks if the user's response to the question is correct
#                outputs reply of chatbot and True if correct, False vice versa
def check_answer(question_tuple, user_input, detected_list):

    question, is_real, obj = question_tuple
    user_answer = user_input.lower()
    object_in_question = obj

    # Extract the number from the user's response (if present)
    user_count = extract_number_from_text(user_answer)

    # Determine the actual count of the object in the image
    actual_count = 0
    actual_id = 0
    for obj_id, obj_label in object_labels.items():
        if obj_id == object_in_question[1]:
            actual_count = detected_list.count(obj_id)
            actual_id = obj_id
            break

    # Check if the object is real or fake
    if is_real:

        if actual_count > 0:

          # One present
          if actual_count == 1:
              # If the object is present, check if the user's count matches the actual count
              if user_count == actual_count:
                  return (f"Correct! There is {actual_count} {object_in_question[0]} in the image.", True)
              else:
                  return (f"Incorrect. There is {actual_count} {object_in_question[0]} in the image.", False)

          # More than one
          else:
              # If the object is present, check if the user's count matches the actual count
              if user_count == actual_count:
                  return (f"Correct! There are {actual_count} {object_labels_plural[actual_id]} in the image.", True)
              else:
                  return (f"Incorrect. There are {actual_count} {object_labels_plural[actual_id]} in the image.", False)

    else:

        # If the object is not present in the image, check if the user correctly stated that
        if "no" in user_answer or "none" in user_answer or user_count == 0 or "aren't" in user_answer:
            return (f"Correct! There are no {object_labels_plural[object_in_question[1]]} in the image.", True)
        else:
            return (f"Incorrect. There are no {object_labels_plural[object_in_question[1]]} in the image, but you said there were.", False)
```

```python
# questions : given detected list, user input and the descrption of the image
#    Outputs: generate 3 question on the detected objects and ask for a description
#             output final mark and errors found.
def questions(user_input, detected_list, desc):
    if user_input[1] == 0:
        print("[Change image: 'stop' -> Stop chatbot: 'quit'] ChatBot: Ok! Let's start with some questions!")

        answers = []
        feedbacks = []
        corrected_answers = []

        # Generate the questions based on the detected objects
        questions_to_ask = generate_questions(detected_list)

        # Ask the generated questions one by one
        for question, is_real, obj in questions_to_ask:
            print(f"[Change image: 'stop' -> Stop chatbot: 'quit'] ChatBot: {question}")
            user_input[0] = input("You: ").lower()

            if user_input[0] == "stop" or user_input[0] == "quit":
              return user_input[0]
            else:
              answers.append(user_input[0])

            corrected_sentence = gf.correct(user_input[0]) # corrects any grammatical errors

            u_temp = user_input[0].translate(str.maketrans('', '', string.punctuation)).split()
            c_temp = (str(corrected_sentence).lower()).translate(str.maketrans('', '', string.punctuation)).split()
            any_err = [c_temp[i] for i in range(0, len(c_temp)) if not c_temp[i] in u_temp]

            if len(any_err) > 0:
              print(f"[Change image: 'stop' -> Stop chatbot: 'quit'] ChatBot: Corrected sentence: {corrected_sentence}")
              #print("any_err ", any_err)
              #print("u_temp ", u_temp)
              #print("c_temp ", c_temp)

            temp_answer = list(corrected_sentence)[0]
            corrected_answers.append(temp_answer)

            # Check the answer and provide feedback
            (feedback, is_right) = check_answer((question, is_real, obj), temp_answer, detected_list)
            feedbacks.append((feedback, is_right))
            print(f"[Change image: 'stop' -> Stop chatbot: 'quit'] ChatBot: {feedback}")

        print("[Change image: 'stop' -> Stop chatbot: 'quit'] ChatBot: Could you describe what's happening in the picture? [max 240 number of characters]")
        user_input[0] = input("You: ").lower()

        if len(user_input[0]) > 240:
          final_answer = (user_input[0])[0:240]
        else:
          final_answer = user_input[0]

        if user_input[0] in ["quit", "stop"]:
            return user_input[0]  # Exit the conversation loop if 'quit' or 'stop' is detected

        # Correct the user's description and print the result
        corrected_description = gf.correct(user_input[0])

        u_temp = user_input[0].translate(str.maketrans('', '', string.punctuation)).split()
        c_temp = str(corrected_description).translate(str.maketrans('', '', string.punctuation)).split()
        any_err = [c_temp[i] for i in range(0, len(c_temp)) if not c_temp[i] in u_temp]

        if len(any_err) > 0:
          print(f"[Change image: 'stop' -> Stop chatbot: 'quit'] ChatBot: Corrected text: {corrected_description}")


        final_mark= calculate_score(questions_to_ask, answers, detected_list, feedbacks, final_answer, corrected_answers, desc)
        print(f"[Change image: 'stop' -> Stop chatbot: 'quit'] ChatBot: Your final mark is: {final_mark}/10")

        # After asking the three questions, offer the user the option to continue or quit
        print("[Change image: 'stop' -> Stop chatbot: 'quit'] ChatBot: Would you like to analyze another image?")
        user_input[0] = input("You: ").lower()

        if user_input[0] in ["quit", "stop"]:
            return user_input[0]

        if classify_input(user_input[0], False):
            return "stop"
        else:
            return "quit"

    return user_input[0]
```

```python
# calculate_score : given all answers and corrected answer, output final mark given based on errors and answer leght
def calculate_score(questions_list, user_answers, detected_list, feedbacks, final_answer, corrected_answers, desc):

    total_score = 0
    max_score_per_question = 5  # Max score for a question is 5 points

    for i, (question) in enumerate(questions_list):
        answer = user_answers[i]
        (_ , is_right) = feedbacks[i]
        corrected_sentence = corrected_answers[i]
        points = 0

        # Check if the answer is correct
        if is_right:

            points += 3  # Correct answer gives 3 points

            # Give additional points if the answer is more detailed (more than a few words)
            if len(answer.split()) > 2:
                points += 1  # Detailed answer gives 1 extra point

            if len(answer.split()) > 4:
                points += 1
        else:
            points -= 1  # Wrong answer loses 1 points
        # Check for grammatical errors
        temp = corrected_sentence#[0]

        # Do not count punctuation as an error
        temp=temp.translate(str.maketrans('', '', string.punctuation))
        answer=answer.translate(str.maketrans('', '', string.punctuation))

        if len(temp)>len(answer):
            res=temp.replace(answer,'')   # Replace equal characters with empty char -> if equal res should be empty
        else:
            res=answer.replace(temp,'')   # Replace equal characters with empty char -> if equal res should be empty

        if len(res)>0:
            points -= 1  # Deduct 1 point for grammatical errors

        # Add points to total score
        total_score += points

    # Normalize total score to a scale of 0 to 10
    max_possible_score = len(questions_list) * max_score_per_question
    final_score = (total_score / max_possible_score) * 10

    # Check final question
    final_points = 0
    desc=desc.split()

    # Do not count stop words of description of the image
    wordsFiltered = [w for w in desc if w not in stop_words]
    unique = []

    if len(final_answer.split()) > 5:  # Reward a more detailed description
        final_points += 6  # Give more points for a good description
    elif len(final_answer.split()) > 0:
        final_points += 4  # Give less points for a basic description

    # Tokenize with NLTK to have POS
    for w in wordsFiltered:
      text = nltk.word_tokenize(w)

      # Get all synonims that are the same POS
      for word, tag in nltk.pos_tag(text):
        # Filter for unique synonyms not equal to word and sort.
        unique_temp = sorted(set(synonym for synonym in get_synonyms(word, tag) if synonym != word))
        unique_temp.append(word)

      # List of synonims
      unique.append(unique_temp)

    words_desc = []
    words_lemma = []
    for w in unique:
      for i in w:
        words_desc.append(i.replace('_',' '))  # Some synonims instead of a space have '_'
        words_lemma.append(lemmatizer.lemmatize(i.replace('_',' ')))

    # Take grammaticaly correct description
    final_corrected = list(gf.correct(final_answer))
    temp_final = final_corrected[0].translate(str.maketrans('', '', string.punctuation))

    present_objects = [(object_labels[obj_id], object_labels_plural[obj_id]) for obj_id in detected_list if obj_id in object_labels]

    # For each user words that appear in the detected or synonim list add a point
    user_desc=temp_final.split()
    for w in user_desc:
      if w in words_desc:
        final_points += 1
      elif lemmatizer.lemmatize(w) in words_lemma:
        final_points += 1
      elif len([True for obj in present_objects if w in obj])>0:
        final_points += 1

    # Check for grammatical errors in the final answer
    final_answer = (final_answer.translate(str.maketrans('', '', string.punctuation))).split()

    # Compare the original and corrected verision of the description
    temp_final=temp_final.split()
    for w in user_desc:
      if w in final_answer:
        final_answer.pop(final_answer.index(w))
        temp_final.pop(temp_final.index(w))

    # Reduce points proportionally to the number of errors detected
    list_errors=final_answer+temp_final
    if len(list_errors) > 0:
        n_error = len(list_errors)/2
        final_points -= n_error  # Deduct 1 point for grammatical errors


    if final_points > 10.0 :
      final_points = 10.0
    elif final_points < 0.0:
      final_points = 0.0

    print(f"[Change image: 'stop' -> Stop chatbot: 'quit'] ChatBot: Your mark for the answering the questions: {round(final_score, 1)}/10")
    print(f"[Change image: 'stop' -> Stop chatbot: 'quit'] ChatBot: Your mark for the description is: {round(final_points, 1)}/10")
    avg = (final_score + final_points) /2

    return round(avg, 1)  # Return score rounded to 1 decimal place
```
